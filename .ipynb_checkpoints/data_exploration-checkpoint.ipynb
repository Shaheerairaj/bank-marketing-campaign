## Data Exploration

```python
df.info()
```

```python
fig, ax = plt.subplots(figsize=(12,10))
df.hist(['age','balance','campaign','pdays','duration'], ax=ax)
plt.show();
```

```python
df_cat_cols = ['job','marital','education','default','housing','loan','contact','month','day','previous']

# Create subplots with two columns
fig, axes = plt.subplots(nrows=len(df_cat_cols), ncols=2, figsize=(16, 5 * len(df_cat_cols)))

# Plot graphs for each column in the first column of the subplot
for i, col in enumerate(df_cat_cols):
    counts = df.groupby([col,'y']).size()
    total_counts = df.groupby(col).size()
    percentages = counts.div(total_counts, level=col).unstack(fill_value=0) * 100
    percentages.plot(kind='bar', stacked=True, ax=axes[i, 0])
    axes[i, 0].set_title(f'Graph for {col}')
    axes[i, 0].set_ylabel('Percentage %')

    # Plot graphs for each column filtered by 'y' == 'yes' in the second column of the subplot
    df[df['y'] == 'yes'].groupby([col,'y']).size().unstack(fill_value=0).plot(kind='bar', stacked=True, ax=axes[i, 1])
    axes[i, 1].set_title(f'Graph for {col} (y = yes)')

# Adjust layout
plt.tight_layout()
plt.show()
```

```python
df['y'].value_counts()/len(df['y'])*100
```

Some interesting findings here:

1. There is an overwhelming majority of people who didn't convert (yes - 11.7% / no - 88.3%)
2. People who are in management or technicians seem to be the ones who convert the most
3. People contacted via cellular were more likely to convert
4. People were drastically more likely to convert the very first time they were contacted
5. There seem to be the most conversions in May, but as a percentage of total people contacted in that month, it doesn't seem to be the highest
6. As a percentage, the highest yielding months were Dcember, October and March
7. Although people in management were the largest number in converts followed by technicians, as a percentage of total people contacted, students and retired people were the most likely to submit a term deposit
8. People without a loan, house or have not been defaulting on their credit are more likely to submit a term deposit

```python
df_num_cols = ['age','balance','campaign','pdays','duration']

for col in df_num_cols:
    if col == 'pdays':
        df[df[col] >= 0][[col,'y']].groupby('y').boxplot()
    else:
        df[[col,'y']].groupby('y').boxplot();
```

Some other interesting findings:

1. People older in age were most likely to subscribe to a term deposit
2. There is a higher likelyhood of the client submitting a term deposit if they have a higher avg yearly balance
3. The less number of days that pass between the previous and current campaign, the more likely the client is to submit a term deposit
4. If the client was contacted for a longer duration in the previous campaign, the more likely they are to submit a term deposit.

### Summary of Findings:

5. There seem to be the most positive responses in May, but as a percentage of total people contacted in that month, it doesn't seem to be the highest. As a percentage, the highest yielding months were Dcember, October and March.
